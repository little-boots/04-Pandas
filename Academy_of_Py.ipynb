```python
# Import libraries
import pandas as pd
```

```python
# Read in data
# NOTE: Course data are kept in (very) separate directory

# File locations
schools_loc  = "../../UNCRAL20190514DATA/02-Homework/04-Pandas/Instructions/PyCitySchools/Resources/schools_complete.csv"
students_loc = "../../UNCRAL20190514DATA/02-Homework/04-Pandas/Instructions/PyCitySchools/Resources/students_complete.csv"

# Data frames
schools_df = pd.read_csv(schools_loc)
students_df = pd.read_csv(students_loc)
```

```python
# Combine all data
all_df = schools_df.merge(students_df, how="outer", on="school_name")

# Make things easier
all_df["pass_math"] = (all_df["math_score"] >= 60)
all_df["pass_reading"] = (all_df["reading_score"] >= 60)
all_df["pass_both"] = (all_df["pass_math"] & all_df["pass_reading"])
```

```python
# Helper code
def formatting(df):
    df["Total Students"] = df["Total Students"].map("{:,}".format)
    df["Total Budget"] = df["Total Budget"].map("${:,}".format)
    df["Average Math Score"] = df["Average Math Score"].map("{:,.1f}".format)
    df["Average Reading Score"] = df["Average Reading Score"].map("{:,.1f}".format)
    df["% Passing Math"] = df["% Passing Math"].map("{:.1%}".format)
    df["% Passing Reading"] = df["% Passing Reading"].map("{:.1%}".format)
    df["Overall Passing Rate"] = df["Overall Passing Rate"].map("{:.1%}".format)
    
recodes = {"size": "Total Students",
           "budget": "Total Budget",
           "math_score": "Average Math Score",
           "reading_score": "Average Reading Score",
           "pass_math": "% Passing Math",
           "pass_reading": "% Passing Reading",
           "pass_both": "Overall Passing Rate"}
```

```python
# School-level
schools_df["Count"] = 1
schools_cols = ["Total Schools", "Total Students", "Total Budget"]
schools_vals = schools_df[["Count", "size", "budget"]].sum().tolist()
schools_data = dict(zip(schools_cols, schools_vals))
schools_summary_df = pd.DataFrame(data=schools_data, index=[0])
schools_summary_df["school"] = "Total"

# Student-level
student_cols = ["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "Overall Passing Rate"]
student_vals = all_df[ ["math_score", "reading_score", "pass_math", "pass_reading", "pass_both"] ].mean().tolist()
student_data = dict(zip(student_cols, student_vals))
student_summary_df = pd.DataFrame(data=student_data, index=[0])
student_summary_df["school"] = "Total"

total_summary_df = schools_summary_df.merge(student_summary_df, how="outer", on="school")

# Still need to clean this up
formatting(total_summary_df)
total_summary_df
```

```python
# School summary
school_summary_df = schools_df[["school_name", "type", "size", "budget"]]
students_grouped = all_df[["school_name", "reading_score", "math_score", "pass_math", "pass_reading", "pass_both"]].groupby("school_name")
school_means = students_grouped.mean()

combine = school_summary_df.merge(school_means, how="outer", on="school_name")
combine = combine.set_index("school_name", drop=True)
combine = combine.rename(columns=recodes)
formatting(combine)
combine
```
